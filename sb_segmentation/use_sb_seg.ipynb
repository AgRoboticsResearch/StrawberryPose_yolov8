Necessary imports for displaying points, boxes, and masks.

```python
import numpy as np
import torch
import matplotlib.pyplot as plt
import cv2
from ultralytics import YOLO
```

```python
image = cv2.imread('/mnt/TempData/brl/pangolin/ruichen/Strawberry_yolov8/strawberry.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,10))
plt.imshow(image)
plt.axis('on')
plt.show()
```

```python
# Load a model
model = YOLO('/mnt/TempData/brl/pangolin/ruichen/Strawberry_yolov8/model/segment.pt')  # load a custom model

# Predict with the model
results = model('/mnt/TempData/brl/pangolin/ruichen/Strawberry_yolov8/strawberry.png')  # predict on an image

# print(results[0].masks)
res = results[0].plot()

plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))   #BGRRGB
```
